```python
#importing data set and displaying
import pandas as pd
df = pd.read_csv('zomato.csv')
display(df.head())
```

```python
#checking for missing values
display(df.isnull().sum())
```

```python
print(len(df))

#this is done to see if its ok to drop 68 rows.
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
df = df.rename(columns={'rate (out of 5)': 'rate','avg cost (two people)': 'avg cost'})
display(df.head())
#renaming the columns
```

```python
df = df.dropna()
print(len(df))
display(df.isnull().sum())
#68 rows for rate column and 57 rows for avg cost column were missing values. so at max about 125 rows would be missing data. which is less compared to the total number of
# rows which was 7105. so i decided dropna would be best because its just negligible amount of rows.
```

```python
df = df.drop_duplicates()
print(len(df))
#no duplicate rows.
```

```python
df['restaurant name'] = df['restaurant name'].str.strip()
df['area'] = df['area'].str.strip().str.title()
df['local address'] = df['local address'].str.strip().str.title()
df['cuisines type'] = df['cuisines type'].str.strip().str.title()
```

```python
display(df.info())
#checking for incorrect data types
```

```python
def rating_category(rate):
    if rate < 3:
        return "Poor"
    elif rate < 3.5:
        return "Average"
    elif rate < 4:
        return "Good"
    else:
        return "Very nice"

df['rating_category'] = df['rate'].apply(rating_category)
#adding this column helps in understanding customer's perception about restaurants.
```

```python
display(df.head())
```

```python
df.to_csv('cleaned_restaurants.csv', index=False)
```
